# How many products cost more than $10.00?

```python
dataset['item_price'] = dataset['item_price'].str.replace('$', '').astype(float)
```

```python
dataset.dtypes
```

```python
dataset['item_name'][dataset['item_price']>10.00].unique()
```

```python
dataset['item_name'][dataset['item_price']>10.00].nunique()
```

# What is the price of each item?

```python
dataset[dataset['quantity']==1]\
    .groupby(by='item_name')['item_price']\
    .first()
```
