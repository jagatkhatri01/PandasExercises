# Which was the most-ordered item?

```python
c = dataset.groupby('item_name')['quantity'].sum().sort_values(ascending=False)
c.head(1)
```

```python
dataset.groupby('item_name')['quantity'].sum().idxmax()
```

#  For the most-ordered item, how many items were ordered?

```python
most_ordered_item = c.index[0]
total_quantity = c.iloc[0]
print(most_ordered_item)
print(total_quantity)
```

# What was the most ordered item in the choice_description column?

```python
dataset.groupby('choice_description')['quantity'].sum().idxmax()
```

```python
cd = dataset.groupby('choice_description')['quantity'].sum().sort_values(ascending=False)
most_choice_description_item = cd.index[0]
total_quantity = cd.iloc[0]
print(most_choice_description_item)
print(total_quantity)
```

# How many items were orderd in total?

```python
total_items_orders = dataset.quantity.sum()
print(total_items_orders)
```

# Turn the item price into a float

# Using Lambda and apply

```python
dataset['item_price'].apply(lambda x: float(x.replace('$', '')))
```

```python
dataset['item_price'] = dataset['item_price'].str.replace('$', '').astype(float)
```

```python
dataset.head(4)
```

```python
dataset['item_price'].dtypes
```

# How much was the revenue for the period in the dataset? 

```python
revenue = (dataset['item_price'] * dataset['quantity']).sum()
print('Revenue was: $' + str(revenue))
```
