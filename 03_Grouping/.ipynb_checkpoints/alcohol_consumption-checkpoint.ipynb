```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
dataset = pd.read_csv(url, sep=',')
```

```python
dataset.shape
```

```python
dataset.head(5)
```

# Which continent drinks more beer on average?

```python
beer_avg = dataset.groupby('continent')['beer_servings'].mean().sort_values(ascending=False)
beer_avg.round(2)
```

```python
top = beer_avg.head(5)
print(top.index[2], top.values[2])
# Can be done using idxmax(), max()
```

```python
last_continent = beer_avg.idxmin()  # index with min value
last_value = beer_avg.min()         # min value

print(last_continent, last_value)
# Using idxmin and min 
# can be done with tail()
```

# For each continent print the statistics for wine consumption.

```python
dataset.groupby('continent').agg({'wine_servings': ['min', 'max', 'sum', 'mean']})
```

```python
dataset.groupby('continent')['wine_servings'].describe()
```

```python
dataset.columns
```

# Print the mean alcohol consumption per continent for every column

```python
dataset.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol']].mean()
```

```python
dataset.groupby('continent').mean(numeric_only=True).round(2)
```

# Print the median alcohol consumption per continent for every column

```python
dataset.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol']].median()
```

```python
# dataset.groupby('continent').agg(['mean','median','min','max']).round(2)
```

# Print the mean, min and max values for spirit consumption.

```python
dataset.groupby('continent').agg({'spirit_servings': ['min', 'max', 'sum', 'mean']})
```

```python
dataset.groupby('continent')['spirit_servings'].mean().sort_values(ascending=False).plot(kind='bar')
```

# Summary

```python
summary = dataset.groupby('continent').agg({
    'beer_servings':['mean','median','min','max'],
    'spirit_servings':['mean','median','min','max'],
    'wine_servings':['mean','median','min','max'],
    'total_litres_of_pure_alcohol':['mean','median','min','max']
}).round(2)

print(summary)
```
